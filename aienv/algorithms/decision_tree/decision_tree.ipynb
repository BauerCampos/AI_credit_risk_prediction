```python
import pickle
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
```

```python
with open('../../database/processed_data/credit.pkl', 'rb') as f:
    x_credit_training, y_credit_training, x_credit_test, y_credit_test = pickle.load(f)
```

```python
arvore = DecisionTreeClassifier(criterion='entropy', random_state = 0)
arvore.fit(x_credit_training, y_credit_training)
```

```python
prediction = arvore.predict(x_credit_test)
prediction
```

```python
y_credit_test
```

```python
precision = accuracy_score(y_credit_test,prediction)
precision
```

```python
cm = ConfusionMatrix(arvore)
cm.fit(x_credit_training,y_credit_training)
cm.score(x_credit_test, y_credit_test)
```

```python
print(classification_report(y_credit_test, prediction))
```
