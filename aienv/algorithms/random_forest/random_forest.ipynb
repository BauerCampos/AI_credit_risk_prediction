```python
import pickle
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
```

```python
with open('../../database/processed_data/credit.pkl', 'rb') as f:
    x_credit_training, y_credit_training, x_credit_test, y_credit_test = pickle.load(f)
```

```python
random_forest = RandomForestClassifier(n_estimators=120, criterion='entropy', random_state=0)
random_forest.fit(x_credit_training, y_credit_training)
```

```python
prediction = random_forest.predict(x_credit_test)
```

```python
accuracy_score(y_credit_test,prediction)
```

```python
cm = ConfusionMatrix(random_forest)
cm.fit(x_credit_training, y_credit_training)
cm.score(x_credit_test, y_credit_test)
```

```python
print(classification_report(y_credit_test, prediction))
```
